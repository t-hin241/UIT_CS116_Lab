#Bước 6: Dùng spatial join (from geopandas.tools import sjoin) để lọc ra các điểm click của người dùng trong 5 quận/huyện hot nhất

```python
merge_data = gpd.sjoin(click_data, top5_dis, how='left', op='within')
```

```python
merge_data
```

```python
merge_data.dropna(axis=0, how='any', subset=['Dist_ID'], inplace=True)
merge_data
```

```python
merge_data = merge_data.sort_values(by='den_growth', axis=0, ascending=False)
merge_data = merge_data.reset_index()
merge_data = merge_data.iloc[:, 2:]
merge_data
```

#Bước 10: Lưu heatmap xuống file png

```python
import numpy as np
from sklearn.cluster import KMeans
from collections import Counter
import pandas as pd
```

```python
dis_top5 = merge_data.Dist_Name.unique().tolist()
dis_top5
```

```python
from folium import plugins
from folium.plugins import HeatMap
m = folium.Map(location=[10.75, 106.7], zoom_start=12, tiles='CartoDB positron')
```

```python
def convert_point(point):
  return np.array([point.x, point.y], dtype = np.float32)

for i in range(len(dis_top5)):
  df = merge_data[merge_data["Dist_Name"] == dis_top5[i]]
  kmean = KMeans(n_clusters=20, random_state=0)
  X = np.array(list(df["geometry"].map(convert_point)))
  kmean.fit(X)

  counter = Counter(kmean.labels_)
  largest_index = np.argmax(counter.values())
  largest_cluster_label = kmean.labels_[largest_index]

  points_largest_cluster = X[kmean.labels_ == largest_index]
  heat_df = pd.DataFrame(points_largest_cluster, columns=['Longitude', 'Latitude'], dtype=np.float32)
  heat_data = [[row['Latitude'], row['Longitude']] for index, row in heat_df.iterrows()]
  HeatMap(heat_data).add_to(m)
  
m
```
