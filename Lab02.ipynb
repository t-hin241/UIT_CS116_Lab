```python
import geopandas as gpd
```

```python
df = gpd.read_file("/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_Ward_Level.shp")
df
```

```python
#Query 1
print("Phường có diện tích lớn nhất là:", 
      df.Com_Name[df.Shape_Area.idxmax()], ",", df.Dist_Name[df.Shape_Area.idxmax()])
print("\n")

#Query 2
print("Phường có dân số 2019 (Pop_2019) cao nhất là:", 
      df.Com_Name[df.Pop_2019.idxmax()], ",", df.Dist_Name[df.Pop_2019.idxmax()])
print("\n")

#Query 3
print("Phường có diện tích nhỏ nhất là:", 
      df.Com_Name[df.Shape_Area.idxmin()], ",", df.Dist_Name[df.Shape_Area.idxmin()])
print("\n")

#Query 4
print("Phường có dân số thấp nhất (2019) là:",
      df.Com_Name[df.Pop_2019.idxmin()], ",", df.Dist_Name[df.Pop_2019.idxmin()])
print("\n")

#Query 5
pop_incr = [pop2019 - pop2009 for pop2019,pop2009 in zip(df.Pop_2019, df.Pop_2009)]
max_pop_incr = pop_incr.index(max(pop_incr))
print("Phường có tốc độ tăng trưởng dân số nhanh nhất là:",
      df.Com_Name[max_pop_incr], ",", df.Dist_Name[max_pop_incr])
print("\n")

#Query 6
min_pop_incr = pop_incr.index(min(pop_incr))
print("Phường có tốc độ tăng trưởng dân số thấp nhất là:",
      df.Com_Name[min_pop_incr], ",", df.Dist_Name[min_pop_incr])
print("\n")

#Query 7
pop_chgs = [abs(pop_incr[i]/df.Pop_2009[i]) if pop_incr[i] > 0 else abs(pop_incr[i]/df.Pop_2019[i]) for i in range(len(pop_incr))]
max_pop_chg = pop_chgs.index(max(pop_chgs))
print("Phường có biến động dân số nhanh nhất là:", 
      df.Com_Name[max_pop_chg], ",", df.Dist_Name[max_pop_chg])
print("\n")

#Query 8
min_pop_chg = pop_chgs.index(min(pop_chgs))
print("Phường có biến động dân số chậm nhất là:", 
      df.Com_Name[min_pop_chg], ",", df.Dist_Name[min_pop_chg])
print("\n")

#Query 9
print("Phường có mật độ dân số cao nhất (2019) là:", 
      df.Com_Name[df.Den_2019.idxmax()], ",", df.Dist_Name[df.Den_2019.idxmax()])
print("\n")

#Query 10
print("Phường có mật độ dân số thấp nhất (2019):", 
      df.Com_Name[df.Den_2019.idxmin()], ",", df.Dist_Name[df.Den_2019.idxmin()])
```
